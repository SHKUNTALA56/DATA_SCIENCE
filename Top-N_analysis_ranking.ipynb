```python
# 1.Order countries based on their water access levels per year.

import pymysql
import pandas as pd

# Create connection
conn = pymysql.connect(
    host='localhost',
    user='root',
    password='Diagnostic_56',
    database='united_nations'
)

# Query
query = "SELECT Country_name, Time_period, Pct_managed_drinking_water_services, ROW_NUMBER() OVER( PARTITION BY Time_period ORDER BY Pct_managed_drinking_water_services ASC) AS Rank_of_water_services FROM united_nations.Access_to_Basic_Services;"
df = pd.read_sql(query, conn)
df
```

```python
# 2. Assess the rankings for countries with the same water level access in a given year.

import pymysql
import pandas as pd

# Create connection
conn = pymysql.connect(
    host='localhost',
    user='root',
    password='Diagnostic_56',
    database='united_nations'
)

# Query
query = "SELECT Country_name, Time_period, Pct_managed_drinking_water_services, ROW_NUMBER() OVER( PARTITION BY Time_period ORDER BY Pct_managed_drinking_water_services ASC) AS Rank_of_water_services FROM united_nations.Access_to_Basic_Services WHERE Pct_managed_drinking_water_services = 100;"
df = pd.read_sql(query, conn)
df
```

```python
# 3. Apply the RANK function instead.

import pymysql
import pandas as pd

# Create connection
conn = pymysql.connect(
    host='localhost',
    user='root',
    password='Diagnostic_56',
    database='united_nations'
)

# Query
query = "SELECT Country_name, Time_period, Pct_managed_drinking_water_services, RANK() OVER( PARTITION BY Time_period ORDER BY Pct_managed_drinking_water_services ASC) AS Rank_of_water_services FROM united_nations.Access_to_Basic_Services;"
df = pd.read_sql(query, conn)
df
```

We can see that in 2015, the country with the lowest percentage of managed drinking water services was Central African Republic . Within the year 2020, we have countries that had the same percentage, such as French Polynesia and Tokelau, and as a result, they have the same rank, 127.
