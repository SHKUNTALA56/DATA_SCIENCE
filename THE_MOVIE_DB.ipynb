```python
import sqlite3
from typing import List, Tuple

def view_table_contents(db_path: str, table_name: str, limit: int = 10) -> List[Tuple]:
   
    try:
        # Connect to the SQLite database
        conn = sqlite3.connect(db_path)
        cursor = conn.cursor()
        
        # Get column names
        cursor.execute(f"PRAGMA table_info({table_name});")
        columns = [column[1] for column in cursor.fetchall()]
        
        # Execute query to retrieve table contents
        cursor.execute(f"SELECT * FROM {table_name} LIMIT {limit};")
        rows = cursor.fetchall()
        
        # Close the connection
        conn.close()
        
        return columns, rows
        
    except sqlite3.Error as e:
        print(f"Database error: {e}")
        if 'conn' in locals() and conn:
            conn.close()
        raise

# View the oscars table
db_path = "TMDB-a-4006 (1).db"
try:
    columns, rows = view_table_contents(db_path, "movies")
    
    # Print column headers
    print(" | ".join(columns))
    print("-" * (sum(len(col) for col in columns) + 3 * (len(columns) - 1)))
    
    # Print rows
    for row in rows:
        print(" | ".join(str(item) for item in row))
        
except Exception as e:
    print(f"Error: {e}")
```

```python
import sqlite3
from typing import List, Tuple, Optional

def view_filtered_table_contents(
    db_path: str, 
    table_name: str, 
    filter_column: str,
    filter_value: any,
    limit: Optional[int] = None
) -> tuple[List[str], List[Tuple]]:
   
    try:
        # Connect to the SQLite database
        conn = sqlite3.connect(db_path)
        cursor = conn.cursor()
        
        # Get column names
        cursor.execute(f"PRAGMA table_info({table_name});")
        columns = [column[1] for column in cursor.fetchall()]
        
        # Execute query to retrieve filtered table contents
        query = f"SELECT * FROM {table_name} WHERE {filter_column} = ?"
        if limit is not None:
            query += f" LIMIT {limit}"
            
        cursor.execute(query, (filter_value,))
        rows = cursor.fetchall()
        
        # Close the connection
        conn.close()
        
        return columns, rows
        
    except sqlite3.Error as e:
        print(f"Database error: {e}")
        if 'conn' in locals() and conn:
            conn.close()
        raise

# View the oscars table filtered for year 2015
db_path = "TMDB-a-4006 (1).db"
try:
    columns, rows = view_filtered_table_contents(
        db_path=db_path,
        table_name="oscars",
        filter_column="year",
        filter_value=2015
    )
    
    # Print column headers
    print(" | ".join(columns))
    print("-" * (sum(len(col) for col in columns) + 3 * (len(columns) - 1)))
    
    # Print rows
    for row in rows:
        print(" | ".join(str(item) for item in row))
        
except Exception as e:
    print(f"Error: {e}")
```

```python
import sqlite3
import pandas as pd

def get_table_structure(db_path: str, table_name: str):
    """
    Get the structure of a table in a SQLite database.
    
    Parameters:
    db_path (str): Path to the SQLite database file
    table_name (str): Name of the table to examine
    
    Returns:
    pandas.DataFrame: DataFrame containing the table structure information
    """
    try:
        # Connect to the SQLite database
        conn = sqlite3.connect(db_path)
        
        # Use pandas to read the PRAGMA table_info command results
        query = f"PRAGMA table_info({table_name});"
        df = pd.read_sql_query(query, conn)
        
        # Close the connection
        conn.close()
        
        # Rename columns for better readability
        column_mapping = {
            'cid': 'Column ID',
            'name': 'Column Name',
            'type': 'Data Type',
            'notnull': 'Not Null',
            'dflt_value': 'Default Value',
            'pk': 'Primary Key'
        }
        df = df.rename(columns=column_mapping)
        
        return df
        
    except sqlite3.Error as e:
        print(f"Database error: {e}")
        if 'conn' in locals() and conn:
            conn.close()
        raise

# Example usage
db_path = "TMDB-a-4006 (1).db"
try:
    # Get the structure of the movies table
    table_name = "genres"
    table_structure = get_table_structure(db_path, table_name)
    
    # Display the table structure
    print(f"Structure of table '{table_name}':")
    display(table_structure)  # In Jupyter notebooks
    
    # If you're not in a Jupyter notebook, use:
    # print(f"Structure of table '{table_name}':")
    # print(table_structure)
    
except Exception as e:
    print(f"Error: {e}")
```

```python
#Question 3 How many unique awards are there in the Oscars table?

import sqlite3
import pandas as pd

def get_unique_values(db_path: str, table_name: str, column_name: str):
    try:
        # Connect to the SQLite database
        conn = sqlite3.connect(db_path)
        
        # Use pandas to read SQL query directly into a DataFrame
        query = f"""
        SELECT DISTINCT {column_name}, COUNT(*) as count
        FROM {table_name}
        GROUP BY {column_name}
        ORDER BY count DESC
        """
        df = pd.read_sql_query(query, conn)
        
        # Close the connection
        conn.close()
        
        return df
        
    except sqlite3.Error as e:
        print(f"Database error: {e}")
        if 'conn' in locals() and conn:
            conn.close()
        raise

# Get unique awards from the oscars table
db_path = "TMDB-a-4006 (1).db"
try:
    # Get unique awards and their counts
    awards_df = get_unique_values(db_path, "oscars", "award")
    
    # Display the results
    print(f"Total number of unique awards: {len(awards_df)}")
    display(awards_df)  # In Jupyter notebooks
    
    # If you're not in a Jupyter notebook, use:
    # print(f"Total number of unique awards: {len(awards_df)}")
    # print(awards_df)
    
except Exception as e:
    print(f"Error: {e}")
```

```python
#Q5. How many movies are there that are both in the "Thriller" genre and contain the word “love” anywhere in the keywords?
import sqlite3
import pandas as pd

def search_movies_by_genre_and_keyword(db_path: str, genre: str, keyword: str):
    try:
        # Connect to the SQLite database
        conn = sqlite3.connect(db_path)
        
        # Use pandas to read SQL query directly into a DataFrame
        query = f"""
        SELECT COUNT(DISTINCT m.movie_id) AS thriller_love_movies
FROM movies m
JOIN genremap gm ON m.movie_id = gm.movie_id
JOIN genres g ON gm.genre_id = g.genre_id
JOIN keywordmap km ON m.movie_id = km.movie_id
JOIN keywords k ON km.keyword_id = k.keyword_id
WHERE g.genre_name = 'Thriller'
  AND LOWER(k.keyword_name) LIKE '%love%';



        """
        df = pd.read_sql_query(query, conn)
        
        # Close the connection
        conn.close()
        
        return df
        
    except sqlite3.Error as e:
        print(f"Database error: {e}")
        if 'conn' in locals() and conn:
            conn.close()
        raise

# Search for Thriller movies with "love" in keywords
db_path = "TMDB-a-4006 (1).db"
try:
    # Get movies that are Thrillers and have "love" in keywords
    thriller_love_movies = search_movies_by_genre_and_keyword(db_path, "Thriller", "love")
    
    # Display the count and the results
    print(f"Number of Thriller movies with 'love' in keywords: {len(thriller_love_movies)}")
    display(thriller_love_movies)  # In Jupyter notebooks
    
    # If you're not in a Jupyter notebook, use:
    # print(f"Number of Thriller movies with 'love' in keywords: {len(thriller_love_movies)}")
    # print(thriller_love_movies)
    
except Exception as e:
    print(f"Error: {e}")
```

```python
# Q9. What are the three production companies that have the highest movie popularity score on average, as recorded within the database?



import sqlite3
import pandas as pd

def search_movies_by_genre_and_keyword(db_path: str, genre: str, keyword: str):
    try:
        # Connect to the SQLite database
        conn = sqlite3.connect(db_path)
        
        # Use pandas to read SQL query directly into a DataFrame
        query = f"""
SELECT pc.production_company_name, AVG(m.popularity) AS avg_popularity
FROM movies m
JOIN productioncompanymap pcm ON m.movie_id = pcm.movie_id
JOIN productioncompanies pc ON pcm.production_company_id = pc.production_company_id
GROUP BY pc.production_company_name
ORDER BY avg_popularity DESC
LIMIT 3;

"""
        df = pd.read_sql_query(query, conn)
        
        # Close the connection
        conn.close()
        
        return df
        
    except sqlite3.Error as e:
        print(f"Database error: {e}")
        if 'conn' in locals() and conn:
            conn.close()
        raise

# Search for Thriller movies with "love" in keywords
db_path = "TMDB-a-4006 (1).db"
try:
    # Get movies that are Thrillers and have "love" in keywords
    thriller_love_movies = search_movies_by_genre_and_keyword(db_path, "Thriller", "love")
    
    # Display the count and the results
    print(f" the three production companies that have the highest movie popularity score on average, as recorded within the database are:{len(thriller_love_movies)}")
    display(thriller_love_movies)  # In Jupyter notebooks
    
    # If you're not in a Jupyter notebook, use:
    # print(f"Number of Thriller movies with 'love' in keywords: {len(thriller_love_movies)}")
    # print(thriller_love_movies)
    
except Exception as e:
    print(f"Error: {e}")
```
